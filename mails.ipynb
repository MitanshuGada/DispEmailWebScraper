```python
import pandas as pd
import os
```

```python
df = pd.read_csv('./yopmail_11_24.csv', index_col=0)
```

```python
df
```

```python
df = df.reset_index()
```

```python
# output_dir = "./emails/Nov 21/"
file_path ='./emails/Nov 24/'
os.makedirs(file_path, exist_ok=True)
```

```python
for index, row in df.iterrows():
    file_loc = row['username']+str(index)+'.html'

    with open(os.path.join(file_path, file_loc), 'w', encoding='utf-8') as file:
        file.write(row['body'])
```
