```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from collections import deque
import os
```

```python
url='https://it.cornell.edu/phish-bowl'
```

```python
options = webdriver.ChromeOptions()
# options.add_argument('--headless')
# browser = webdriver.Chrome(options=options)
```

```python
df = pd.DataFrame(columns=['title', 'date_time', 'email_message'])
```

```python
browser = webdriver.Chrome(options=options)
# emails_per_page = 15
# df = pd.dataFrame()
data = {'title': [],
        'date_time': [],
        'email_message': []}

browser.get(url)
for j in range(0, 9):
    print("Page: ",j)

    for i in range(1, 21):
        title = browser.find_element(By.CSS_SELECTOR, "tr:nth-child("+str(i)+") a").text
        browser.find_element(By.CSS_SELECTOR, "tr:nth-child("+str(i)+") a").click()
        date_time = browser.find_element(By.CSS_SELECTOR, "time").text
        # print(date_time)
        # print('-'*100)
        # print(browser.page_source)
        email_message = browser.find_element(By.CSS_SELECTOR, ".body-container-emails").get_attribute('innerHTML')
        # print(email_message)
        # print('-'*100)
        # print(browser.page_source)
        browser.back()
        print(title)
        data["date_time"].append(date_time)
        data['title'].append(title)
        data['email_message'].append(email_message)
    print('-'*100)
    browser.find_element(By.CSS_SELECTOR,".pager__item--next > a").click()
    # print(browser.page_source)

    # break
```

```python
df=pd.DataFrame(data)
```
