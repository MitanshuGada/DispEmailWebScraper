## Code to check number of scraped emails from csv files

```python
import os
import glob
import pandas as pd
```

```python
path = "./"
extension = "csv"
os.chdir(path=path)
```

```python
listOfFileNames= glob.glob('yopmail*.{}'.format(extension))
```

```python
listOfFileNames
```

```python
tempData = {'fileName' : [], 'numOfEmails': []}
```

```python
for fileName in listOfFileNames:
    df = pd.read_csv(fileName, index_col=0)
    tempData['fileName'].append(fileName)
    tempData['numOfEmails'].append(df.size)
    print(f"{fileName} => {df.size}")
```

```python
sum(tempData['numOfEmails'])
```
