```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from collections import deque
import os
```

```python
url="https://stats.security.gatech.edu/phishbowl/?notice="
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
browser = webdriver.Chrome(options=options)
```

```python
# browser = webdriver.Chrome(options=options)
# emails_per_page = 15
# df = pd.dataFrame()
data = {'date': [],
        'description': [],
        'image_url': [],
        'label': []}
err = []
for i in range(1,235):
    
    browser.get(url+str(i))
    print(f"{i}...")
    try:
        data['date'].append(browser.find_element(By.CSS_SELECTOR, ".date").text)
        try:
            # phishing
            data['label'].append(" ".join(browser.find_element(By.CSS_SELECTOR, ".mtitle1").text.split(' ')[-2:]))
        except:
            # safe
            data['label'].append(" ".join(browser.find_element(By.CSS_SELECTOR, ".mtitle0").text.split(' ')[-2:]))


        data['description'].append(browser.find_element(By.CSS_SELECTOR, ".desc").text)
        data['image_url'].append(browser.find_element(By.CSS_SELECTOR, ".msg_img").get_attribute('src'))
    except:
        print(f"{i} err {'-'*10}")
        err.append(url)
    # print(data)
    print(f"{i} done")
    # break
```

```python
df = pd.DataFrame(data)
```
