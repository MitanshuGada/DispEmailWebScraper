```python
def yopmail(url="", usernames=[]):
    browser = webdriver.Chrome(options=options)
    emails_per_page = 15

    # num_mails = browser.find_element(By.ID, "nbmail").text.split()[0]
    # num_pages = ceil(num_mails // emails_per_page + 1)
    yopmail_dict = {'url': [],
                    'username': [],
                    'sender': [],
                    'date_time': [],
                    'subject': [],
                    'body': [],
                }
    
    tempQueue = deque(usernames.copy())
    while(len(tempQueue) != 0):
        user = tempQueue.popleft()
        browser.get(url+user)
        
        try:
            num_mails = int(browser.find_element(By.ID, "nbmail").text.split()[0])
            print(f"{num_mails=}")
            if (num_mails==-1):
                raise Exception("no mails try again")
        except:
            print(url+user)
            tempQueue.append(user)
            time.sleep(30)
            continue

            
        
        num_pages = (num_mails // emails_per_page) + 1 
        print(f"{num_pages=}")
    
        count =1
        for j in range(1, num_pages+1):
            # click on the first mail
            for i in range(1, emails_per_page+1):
                count+=1
                try:
                    browser.switch_to.frame(0)

                    browser.find_element(By.XPATH, "//div["+ str(i+1) + "]/button/div/span[2]").click()
                    browser.switch_to.default_content()
                
                    # switch to mail body
                    browser.switch_to.frame(2)
                
                    #subject
                    subject = browser.find_element(By.CSS_SELECTOR, ".ellipsis:nth-child(1)").text
                
                    # sender
                    sender = browser.find_element(By.CSS_SELECTOR, ".b:nth-child(2)").text
        
                    # time
                    dateTime = browser.find_element(By.CSS_SELECTOR, ".md:nth-child(3) > .ellipsis").text
        
                    #body
                    body = browser.find_element(By.XPATH, "//html/body/main[@class='yscrollbar']").get_attribute('innerHTML')
                    
                    browser.switch_to.default_content()

                    # saving the content to a dictionary
                    yopmail_dict['url'].append(url)
                    yopmail_dict['username'].append(user)
                    yopmail_dict['sender'].append(sender)
                    yopmail_dict['date_time'].append(dateTime)
                    yopmail_dict['subject'].append(subject)
                    yopmail_dict['body'].append(body)
                    print(count)

                except:
                    continue
            try:
                # change the page.
                browser.switch_to.default_content()
                browser.find_element(By.CSS_SELECTOR, '#next .material-icons-outlined').click()
                browser.switch_to.default_content()
            except:
                continue            

    return yopmail_dict 
```

```python
yopmail_out = yopmail("https://yopmail.com/?", usernames)
```
